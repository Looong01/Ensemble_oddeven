```python
def val_epoch_meta(model1, model2, model3, val_loader, device):
    model1.train()
    model2.train()
    model3.train()

    num_iter = len(val_loader)
    valloss = 0
    all_labels1 = []
    all_preds1 = []

    all_labels2 = []
    all_preds2 = []

    all_labels3 = []
    all_preds3 = []
    
    all_even_num1 = []
    all_even_num2 = []

    model1.eval()
    model2.eval()
    model3.eval()

    for i, (img1, img2, img1_label, img2_label, meta_label, even_num1, even_num2) in enumerate(val_loader):
        #传入cuda中
        img1, img2, img1_label, img2_label, meta_label = img1.to(device), img2.to(device), img1_label.to(device), img2_label.to(device), meta_label.to(device)
        with torch.no_grad():
            feature1 = model1(img1)#select features
            feature2 = model1(img2)#select features

            combine_features = torch.cat((feature1, feature2), dim=1)#to model 3


            label_pred1 = model2(feature1).squeeze()
            label_pred2 = model2(feature2).squeeze()

            label_pred3 = model3(combine_features).squeeze()

            #二分类损伤函数F.binary_cross_entropy_with_logits
            loss1 = F.binary_cross_entropy_with_logits(label_pred1, img1_label.float())
            loss2 = F.binary_cross_entropy_with_logits(label_pred2, img2_label.float())
            loss3 = F.binary_cross_entropy_with_logits(label_pred3, meta_label.float())
            img_loss = (loss1+loss2)/2.0
            combine_loss = img_loss + loss3
            valloss += combine_loss
            #acc
            preds1 = torch.sigmoid(label_pred1)
            preds1 = (preds1 > 0.5).float()

            preds2 = torch.sigmoid(label_pred2)
            preds2 = (preds2 > 0.5).float()

            preds3 = torch.sigmoid(label_pred3)
            preds3 = (preds3 > 0.5).float()

            all_labels1.extend(img1_label.cpu().numpy())
            all_preds1.extend(preds1.cpu().numpy())

            all_labels2.extend(img2_label.cpu().numpy())
            all_preds2.extend(preds2.cpu().numpy())

            all_labels3.extend(meta_label.cpu().numpy())
            all_preds3.extend(preds3.cpu().numpy())
            
            all_even_num1.extend(even_num1)
            all_even_num2.extend(even_num2)

    valloss = valloss / num_iter
    return all_labels1,all_preds1, all_labels2, all_preds2, all_labels3, all_preds3, all_even_num1, all_even_num2
```

```python
def psychometric_curve(x,alpha,beta):
    return 1. / (1 + np.exp( -(x - alpha) / beta ))

def plot_psychometric_curve(df,):
    from scipy import optimize
    df['response'] = np.array(df['preds'].values,dtype = int)
    df['answer'] = np.array(df['labels'].values,dtype = int)
    # fit psychometric curve
    # fit psychometric curve
    df_temp = dict(n_even = [],
                   true_prob = [],
                   pred_prob = [],
                   )
    # fit psychometric curve
    df_temp = dict(n_even = [],
                   true_prob = [],
                   pred_prob = [],
                   )
    for (n_even),df_sub in df.groupby('stimulus'):
        df_temp['n_even'].append(n_even)
        df_temp['true_prob'].append(df_sub['answer'].mean())
        df_temp['pred_prob'].append(df_sub['response'].mean())
    df_temp = pd.DataFrame(df_temp)
    
    df_curve = dict(x=[],y=[])
    popt, pcov = optimize.curve_fit(psychometric_curve, 
                                    df_temp['n_even'].values.astype(float) - df_temp['n_even'].values.astype(float).min(), 
                                    df_temp['pred_prob'].values, 
                                    method="lm")
    x = np.linspace(0,
                    df_temp['n_even'].values.astype(float).max() - df_temp['n_even'].values.astype(float).min(),
                    1000)
    y = psychometric_curve(x,*popt)
    [df_curve['x'].append(item) for item in x]
    [df_curve['y'].append(item) for item in y]
    df_curve = pd.DataFrame(df_curve)
    return df_curve,df
```

```python
df_curve,df = plot_psychometric_curve(df)
```
